# Resampling  
Dataset Link: https://www.kaggle.com/datasets/crischir/sample-wav-audio-files

```python
import librosa
audio_path =" /kaggle/input/sample-wav-audio-files/BAK.wav"
```

#### Loads the audio file into Python
#### Returns two things:

- signal: the audio waveform as a NumPy array (like a list of numbers representing the sound amplitude over time)
- fs_original: the original sampling rate of the audio (number of samples per second, in Hz)
- sr=None: tells Librosa not to change the original sampling rate

```python
# load audio
signal, fs_original = librosa.load(audio_path, sr=None)
```

```python
# new sampling rate
fs_new = 16000
```

```python
# resampling
resampled = librosa.resample(
    signal,
    orig_sr=fs_original,
    target_sr=fs_new
)
```

```python
print("Original Sampling Rate:", fs_original)
print("Original Number of Samples:", len(signal))
print("New Sampling Rate:", fs_new)
print("Resampled Number of Samples:", len(resampled))
```
