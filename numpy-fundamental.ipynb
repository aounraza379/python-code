# NumPy Basics

This notebook covers:
- Creating NumPy arrays
- Array shapes and reshaping
- Element-wise operations
- Broadcasting
- Matrix multiplication
- Indexing, slicing, and masking


```python
import numpy as np
```

```python
# 1D array (vector)
arr_1d = np.array([1, 2, 3, 4])
print("1D Array:")
print(arr_1d)

# 2D array (matrix)
arr_2d = np.array([[10, 20], [30, 40]])
print("\n2D Array:")
print(arr_2d)

# Array of zeros
arr_zeros = np.zeros(5)
print("\nZeros Array:")
print(arr_zeros)

# Linspace array
arr_lin = np.linspace(0, 1, 3)
print("\nLinspace Array:")
print(arr_lin)
```

```python
# Array Attributes & Reshaping
data = np.arange(12, dtype=np.int32)
print("Original Data:", data)
print("Shape:", data.shape)
print("Dtype:", data.dtype)

matrix = data.reshape(3, 4)
print("\nReshaped Matrix:")
print(matrix)
print("Matrix Shape:", matrix.shape)
print("Matrix Dtype:", matrix.dtype)
```

```python
# Element-Wise Operations
a = np.array([1, 2, 3])
b = np.array([10, 20, 30])

print("a + b =", a + b)
print("a * 2 =", a * 2)
print("a * b =", a * b)
```

```python
# Broadcasting
A = np.array([[1, 2, 3],
              [4, 5, 6]])

scalar = 10

print("A + 10:")
print(A + scalar)
```

```python
# Matrix Multiplication vs Element-Wise
X = np.array([[1, 2],
              [3, 4]])

Y = np.array([[5, 6],
              [7, 8]])

print("Element-wise multiplication (X * Y):")
print(X * Y)

print("\nMatrix multiplication (X @ Y):")
print(X @ Y)
```

```python
# Boolean Masking
mask = matrix > 20
print("Boolean Mask:")
print(mask)

print("\nFiltered values (>20):")
print(matrix[mask])
```
